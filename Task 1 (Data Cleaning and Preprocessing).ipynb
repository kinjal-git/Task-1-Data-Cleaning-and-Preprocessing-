```python
#Standardizing the text values
df['Education'].unique()
```

```python
df['Marital_Status'].unique()
```

```python
ed_map = { 'Graduation':'Graduation', 'PhD':'Higher', 'Master':'Higher', 'Basic':'Basic', '2n Cycle':'Higher'}
df['Education'] = df['Education'].map(ed_map)
print(ed_map)
```

```python
ma_map = {'Single':1, 'Together':2, 'Married':2, 'Divorced':1, 'Widow':1, 'Alone':1,'Absurd':1, 'YOLO':1}
df['Marital_Status']=df['Marital_Status'].map(ma_map)
print(ma_map)
```

```python
#Converting date formats to a consistent type
df['Dt_Customer']=pd.to_datetime(df['Dt_Customer'],format='%d-%m-%Y')
print(df['Dt_Customer'])
```

```python
df['Age']=2025-df['Year_Birth']
print(df['Age'])
```

```python
df['Tenure (Days)']=(pd.to_datetime('today')-df['Dt_Customer']).dt.days
print(df['Tenure (Days)'])
```

```python
#Renaming column headers to be clean and uniform 
current_headers = list(df.columns.values)
print(current_headers)
```

```python
clean_headers = {c: c.strip() for c in df.columns}
print(clean_headers)
```

```python
#Checking and fixing data types
df.dtypes
```

```python
#Data Transformation
df['Total_Spending'] = df[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].sum(axis=1)
print(df['Total_Spending'])
```

```python
df['Family Size']= df['Kidhome'] + df['Teenhome']+ df['Marital_Status']
print(df['Family Size'])
```

```python
df=df.drop(['Z_CostContact', 'Z_Revenue','ID', 'Year_Birth','Dt_Customer'],axis=1)
```
